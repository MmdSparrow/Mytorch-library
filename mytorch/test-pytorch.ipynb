```python
from torch import tensor
import torch
import torch.nn as nn

a=tensor([[[[1,3,1],[3,1,2]], [[1,3,1],[3,1,2]]], [[[2,2,2],[2,2,2]], [[1,3,1],[3,1,2]]]], dtype=torch.float32)
conv_layer = nn.Conv2d(in_channels=2, out_channels=1, kernel_size=2, stride=1, padding=0)

# Perform the convolution operation
output = conv_layer(a)

print(output)

```
